```python
def edit(temp_data,temp_acc_no):
    while True:
        cls()
        print("Enter 1.Credit\nEnter 2.Debit\nEnter 3.Check Balance\n 4.Logout")
        ch = int(input("Enter choice: "))
        global b
        if ch == 1:
            temp_data['bal']+=int(input("Enter Amount to Credit: "))
            print(temp_data)
            fp = open(f"./Bank_data/{str(temp_acc_no)}",'w')
            json.dump(temp_data,fp)
            fp.close()
            print("\n\n Amount has been successfully credited to ur account\n\n")
            input("Press Enter key")
        elif ch == 2:
            temp_amount = int(input("Enter Amount to Debit: "))
            if temp_amount<temp_data['bal']:
                temp_data['bal']-=temp_amount
                fp = open(f"./Bank_data/{str(temp_acc_no)}",'w')
                json.dump(temp_data,fp)
                fp.close()
                print("\n\n Amount has been successfully debited from ur account\n\n")
            else:
                print("Your current balance is low.Please try lower amount. ")
            input("..Press Enter Key..")
        elif ch == 3:
            fp = open(f"./Bank_data/{str(temp_acc_no)}",'r')
            temp_databal = json.load(fp)
            print("Your current account balance: ",temp_databal['bal'])
            input("..Press Enter Key..")
        elif ch == 4:
            print("\n\n")
            print("Logging out..")
            time.sleep(2)
            return 0
        else:
            print("\n\n Entered Wrong Choice")
            
         
                
```
